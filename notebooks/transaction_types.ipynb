### Transaction Types
1. Instant Transaction: Inter-node transaction with no duration attribute
2. Individual Span Transaction: Single node transaction with duration attribute.
3. Inter-node span Transaction: Inter-node transaction with duration attribute.

```python
import pandas as pd
import plotly.graph_objects as go
import datetime
```

```python
def add_time(dt, duration):
    return (dt+ datetime.timedelta(0,86400*int(duration)))
```

#### Visualizing Instant Transaction

```python
df = pd.read_csv('../data/SAMPLE/Calls.csv')
df.Time = pd.to_datetime(df.Time)
```

```python
tim_arr, y_arr = [], []
for i in range(len(df.Time)):
    tim_arr.append(df.Time[i])
    tim_arr.append(df.Time[i])
    tim_arr.append(df.Time[i])
    y_arr.append(df.Source[i])
    y_arr.append(df.Target[i])
    y_arr.append(None)
fig = go.Figure()
fig.add_trace(go.Scatter(x = tim_arr, y = y_arr, opacity = 0.5,legendgroup="Calls", name='Calls', line=dict(color='brown', width=1)))
fig.add_trace(go.Scatter(x=df['Time'], y= df['Source'],opacity = 0.5, legendgroup="Calls", name='Source', mode='markers', marker_symbol=100, marker=dict(color='brown',size=5)))
fig.add_trace(go.Scatter(x=df['Time'], y= df['Target'],opacity = 0.5, legendgroup="Calls", name='Target', mode='markers', marker_symbol=11, marker=dict(color='brown',size=5)))
fig.update_layout(
    xaxis = dict(
        side='top'
    ),
    height=600
)
fig.show()
```

#### Visualizing Individual Span Transaction

```python
df2 = pd.read_csv('../data/travel/Travel.csv')
df2.Time = pd.to_datetime(df2.Time)
```

```python
tim_arr2, time_list, y2 = [], [], []
for i in range(len(df2['Source'])):
    tim_arr2.append(df2['Time'][i])
    temp = add_time(df2['Time'][i], df2['Duration'][i])
    tim_arr2.append(temp)
    time_list.append(temp)
    tim_arr2.append(None)
    y2.append(df2['Source'][i])
    y2.append(df2['Source'][i])
    y2.append(None)
fig1 = go.Figure()

fig1.add_trace(go.Scatter(x = tim_arr2, y = y2,opacity = 0.5, legendgroup="Travel", name='Travel', line=dict(color='maroon', width=3)))
fig1.add_trace(go.Scatter(x=df2['Time'],opacity = 0.8, y= df2['Source'], legendgroup="Travel", name='Source', mode='markers', marker_symbol=0, marker=dict(color='maroon',size=5)))
fig1.add_trace(go.Scatter(x=time_list,opacity = 0.8, y= df2['Source'], legendgroup="Travel", name='Destination', mode='markers', marker_symbol=11, marker=dict(color='maroon',size=5)))

fig1.update_layout(
    xaxis = dict(
        side='top'
    ),
    height=600
)
fig1.show()
```

#### Visualizing Internode Span Transaction

```python
df3 = pd.read_csv('../data/post/Post.csv')
df3.Time = pd.to_datetime(df3.Time)
```

```python
tim_arr3, y_arr3, time_list3 = [], [], []
for i in range(len(df3.Time)):
    tim_arr3.append(df3.Time[i])
    temp = add_time(df3['Time'][i], df3['Duration'][i])
    tim_arr3.append(temp)
    time_list3.append(temp)
    tim_arr3.append(df3.Time[i])
    y_arr3.append(df3.Source[i])
    y_arr3.append(df3.Target[i])
    y_arr3.append(None)
fig3 = go.Figure()
fig3.add_trace(go.Scatter(x = tim_arr3, y = y_arr3, opacity = 0.5,legendgroup="Post", name='Post', line=dict(color='blue', width=2)))
fig3.add_trace(go.Scatter(x=df3['Time'], y= df3['Source'],opacity = 0.5, legendgroup="Calls", name='Source', mode='markers', marker_symbol=0, marker=dict(color='blue',size=5)))
fig3.add_trace(go.Scatter(x=time_list3, y= df3['Target'],opacity = 0.5, legendgroup="Calls", name='Target', mode='markers', marker_symbol=11, marker=dict(color='blue',size=5)))
fig3.update_layout(
    xaxis = dict(
        side='top'
    ),
    height=600
)
fig3.show()
```
